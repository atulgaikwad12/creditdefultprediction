## Dataset Information

This dataset contains information on default payments, demographic factors, credit data, history of payment, and bill statements of credit card clients in Taiwan from April 2005 to September 2005.

Content
There are 25 variables:

1. ID: ID of each client
2. LIMIT_BAL: Amount of given credit in NT dollars (includes individual and family/supplementary credit
3. SEX: Gender (1=male, 2=female)
4. EDUCATION: (1=graduate school, 2=university, 3=high school, 4=others, 5=unknown, 6=unknown)
5. MARRIAGE: Marital status (1=married, 2=single, 3=others)
6. AGE: Age in years
7. PAY_0: Repayment status in September, 2005 (-1=pay duly, 1=payment delay for one month, 2=payment delay for two months, â€¦ 8=payment delay for eight months, 9=payment delay for nine months and above)
PAY_2: Repayment status in August, 2005 (scale same as above)
PAY_3: Repayment status in July, 2005 (scale same as above)
PAY_4: Repayment status in June, 2005 (scale same as above)
PAY_5: Repayment status in May, 2005 (scale same as above)
PAY_6: Repayment status in April, 2005 (scale same as above)
8. BILL_AMT1: Amount of bill statement in September, 2005 (NT dollar)
BILL_AMT2: Amount of bill statement in August, 2005 (NT dollar)
BILL_AMT3: Amount of bill statement in July, 2005 (NT dollar)
BILL_AMT4: Amount of bill statement in June, 2005 (NT dollar)
BILL_AMT5: Amount of bill statement in May, 2005 (NT dollar)
BILL_AMT6: Amount of bill statement in April, 2005 (NT dollar)
9. PAY_AMT1: Amount of previous payment in September, 2005 (NT dollar)
PAY_AMT2: Amount of previous payment in August, 2005 (NT dollar)
PAY_AMT3: Amount of previous payment in July, 2005 (NT dollar)
PAY_AMT4: Amount of previous payment in June, 2005 (NT dollar)
PAY_AMT5: Amount of previous payment in May, 2005 (NT dollar)
PAY_AMT6: Amount of previous payment in April, 2005 (NT dollar)
10. default.payment.next.month: Default payment (1=yes, 0=no)

## Notes/Observations: 

1. As per Data this is binary classification problem where "default.payment.next.month" is label column so can be solved using supervised machine learning algorithams.

2. ID column had no significance so will remove it from datasset as it is just like another index.

3. Dataset has no duplicate row but after excluding ID column it has 35 duplicate rows though for this kind of data seems this duplicates enteries are valid(possiblity of having more than one cardholders with same limit balance and payment history,billing history) so not removing them for now.

4. all categorical variables having int/float data type so no need of encoding techqniues (like label ordinal encoding/ one hot enconding/ target guided ordinal/ mean encoding )

5. No missing values but has zero values in some columns pay_0 - pay_6, BILL_AMT1 - BILL_AMT6. Here bill amount =0 is fine but for columns pay_0 - pay_6 significance of 0 not defined so leaving it as it is as replacing/removing it may cause loss of info.

6. pay_0 to pay_6 are highly correlated to each other and to BILL_AMT1 - BILL_AMT6 but this seems valid considering business usecase (Payment of 1st month affects subsequent payments of next months same way it affects BillAmts of subsequent months). For keeping as it is but for linear regression model still we can experiment with new derived column replacing pay_0 to pay_6 columns with one single column to try reducing multicollinearity.

7. sex, education , marriage, age, BILL_AMT1 - BILL_AMT6 seems having low impact directly on default.payment.next.month but indirectly somewhere they are impacting Limit Bal,pay_0-pay-6 which then highly correlated with label column(default.payment.next.month).

8. Will use standardization (for linear regression/logistic/knn algos) as different scales values exists in independant variables.
